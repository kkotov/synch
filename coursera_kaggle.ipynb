```python
train_df["date"] = pd.to_datetime(train_df["date"])
```

```python
tmp = pd.merge(train_df, items_df, on="item_id")
```

```python
tmp.tail()
```

```python
cum_sales = {shop_id: shop_df.date.value_counts() for (shop_id, shop_df) in tmp.groupby("shop_id")}
#sales = tmp.loc[lambda tmp: tmp.shop_id == 2, ["date","shop_id"]].groupby("date").count()

plt.figure()
cum_sales[2].plot()

#sales[2].groupby(by=[b.index.month, b.index.year])
cum_sales_rough = {store_id: s.groupby(pd.Grouper(freq='M')).sum() for (store_id, s) in cum_sales.iteritems()}

cum_sales_rough[2].plot()
```
