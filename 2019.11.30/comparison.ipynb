```python
def match(row):
    target_lag_1 = (row['item_cnt_month_lag_1'] - row['target_lag_1'])/row['target_lag_1'] if row['target_lag_1']!=0 else row['item_cnt_month_lag_1']
    target_lag_2 = (row['item_cnt_month_lag_2'] - row['target_lag_2'])/row['target_lag_2'] if row['target_lag_2']!=0 else row['item_cnt_month_lag_2']
    target_lag_3 = (row['item_cnt_month_lag_3'] - row['target_lag_3'])/row['target_lag_3'] if row['target_lag_3']!=0 else row['item_cnt_month_lag_3']
    target_lag_6 = (row['item_cnt_month_lag_6'] - row['target_lag_6'])/row['target_lag_6'] if row['target_lag_6']!=0 else row['item_cnt_month_lag_6']
    target_lag_12= (row['item_cnt_month_lag_12'] -row['target_lag_12'])/row['target_lag_12'] if row['target_lag_12']!=0 else row['item_cnt_month_lag_12']
    total_monthly_sales_lag_1 = (row['date_avg_item_cnt_lag_1'] - row['total_monthly_sales_lag_1'])/row['total_monthly_sales_lag_1'] if row['total_monthly_sales_lag_1']!=0 else row['date_avg_item_cnt_lag_1']
    total_monthly_item_sales_lag_1 = (row['date_item_avg_item_cnt_lag_1'] - row['total_monthly_item_sales_lag_1'])/row['total_monthly_item_sales_lag_1'] if row['total_monthly_item_sales_lag_1']!=0 else row['date_item_avg_item_cnt_lag_1']
    total_monthly_item_sales_lag_2 = (row['date_item_avg_item_cnt_lag_2'] - row['total_monthly_item_sales_lag_2'])/row['total_monthly_item_sales_lag_2'] if row['total_monthly_item_sales_lag_2']!=0 else row['date_item_avg_item_cnt_lag_2']
    total_monthly_item_sales_lag_3 = (row['date_item_avg_item_cnt_lag_3'] - row['total_monthly_item_sales_lag_3'])/row['total_monthly_item_sales_lag_3'] if row['total_monthly_item_sales_lag_3']!=0 else row['date_item_avg_item_cnt_lag_3']
    total_monthly_item_sales_lag_6 = (row['date_item_avg_item_cnt_lag_6'] - row['total_monthly_item_sales_lag_6'])/row['total_monthly_item_sales_lag_6'] if row['total_monthly_item_sales_lag_6']!=0 else row['date_item_avg_item_cnt_lag_6']
    total_monthly_item_sales_lag_12 = (row['date_item_avg_item_cnt_lag_12'] - row['total_monthly_item_sales_lag_12'])/row['total_monthly_item_sales_lag_12'] if row['total_monthly_item_sales_lag_12']!=0 else row['date_item_avg_item_cnt_lag_12']
    total_monthly_shop_sales_lag_1 = (row['date_shop_avg_item_cnt_lag_1'] - row['total_monthly_shop_sales_lag_1'])/row['total_monthly_shop_sales_lag_1'] if row['total_monthly_shop_sales_lag_1']!=0 else row['date_shop_avg_item_cnt_lag_1']
    total_monthly_shop_sales_lag_2 = (row['date_shop_avg_item_cnt_lag_2'] - row['total_monthly_shop_sales_lag_2'])/row['total_monthly_shop_sales_lag_2'] if row['total_monthly_shop_sales_lag_2']!=0 else row['date_shop_avg_item_cnt_lag_2']
    total_monthly_shop_sales_lag_3 = (row['date_shop_avg_item_cnt_lag_3'] - row['total_monthly_shop_sales_lag_3'])/row['total_monthly_shop_sales_lag_3'] if row['total_monthly_shop_sales_lag_3']!=0 else row['date_shop_avg_item_cnt_lag_3']
    total_monthly_shop_sales_lag_6 = (row['date_shop_avg_item_cnt_lag_6'] - row['total_monthly_shop_sales_lag_6'])/row['total_monthly_shop_sales_lag_6'] if row['total_monthly_shop_sales_lag_6']!=0 else row['date_shop_avg_item_cnt_lag_6']
    total_monthly_shop_sales_lag_12 = (row['date_shop_avg_item_cnt_lag_12'] - row['total_monthly_shop_sales_lag_12'])/row['total_monthly_shop_sales_lag_12'] if row['total_monthly_shop_sales_lag_12']!=0 else row['date_shop_avg_item_cnt_lag_12']
    total_monthly_category_sales_lag_1 = (row['date_cat_avg_item_cnt_lag_1'] - row['total_monthly_category_sales_lag_1'])/row['total_monthly_category_sales_lag_1'] if row['total_monthly_category_sales_lag_1']!=0 else row['date_cat_avg_item_cnt_lag_1']
    total_monthly_shop_category_sales_lag_1 = (row['date_shop_cat_avg_item_cnt_lag_1'] - row['total_monthly_shop_category_sales_lag_1'])/row['total_monthly_shop_category_sales_lag_1'] if row['total_monthly_shop_category_sales_lag_1']!=0 else row['date_shop_cat_avg_item_cnt_lag_1']
    total_monthly_shop_supercategory_sales_lag_1 = (row['date_shop_type_avg_item_cnt_lag_1'] - row['total_monthly_shop_supercategory_sales_lag_1'])/row['total_monthly_shop_supercategory_sales_lag_1'] if row['total_monthly_shop_supercategory_sales_lag_1']!=0 else row['date_shop_cat_avg_item_cnt_lag_1']
    total_monthly_shop_subcategory_sales_lag_1 = (row['date_shop_subtype_avg_item_cnt_lag_1'] - row['total_monthly_shop_subcategory_sales_lag_1'])/row['total_monthly_shop_subcategory_sales_lag_1'] if row['total_monthly_shop_subcategory_sales_lag_1']!=0 else row['total_monthly_shop_subcategory_sales_lag_1']
    total_monthly_city_sales_lag_1 = (row['date_city_avg_item_cnt_lag_1'] - row['total_monthly_city_sales_lag_1'])/row['total_monthly_city_sales_lag_1'] if row['total_monthly_city_sales_lag_1']!=0 else row['date_city_avg_item_cnt_lag_1']
    total_monthly_city_item_sales_lag_1 = (row['date_item_city_avg_item_cnt_lag_1'] - row['total_monthly_city_item_sales_lag_1'])/row['total_monthly_city_item_sales_lag_1'] if row['total_monthly_city_item_sales_lag_1']!=0 else row['date_item_city_avg_item_cnt_lag_1']
    total_monthly_supercategory_sales_lag_1 = (row['date_type_avg_item_cnt_lag_1'] - row['total_monthly_supercategory_sales_lag_1'])/row['total_monthly_supercategory_sales_lag_1'] if row['total_monthly_supercategory_sales_lag_1']!=0 else row['date_type_avg_item_cnt_lag_1']
    total_monthly_subcategory_sales_lag_1 = (row['date_subtype_avg_item_cnt_lag_1'] - row['total_monthly_subcategory_sales_lag_1'])/row['total_monthly_subcategory_sales_lag_1'] if row['total_monthly_subcategory_sales_lag_1']!=0 else row['date_subtype_avg_item_cnt_lag_1']

    return \
    target_lag_1 + target_lag_2 + target_lag_3 + target_lag_6 + target_lag_12 + \
        total_monthly_sales_lag_1 + \
        total_monthly_item_sales_lag_1 + total_monthly_item_sales_lag_2 + total_monthly_item_sales_lag_3 + \
        total_monthly_item_sales_lag_6 + total_monthly_item_sales_lag_12 + \
        total_monthly_shop_sales_lag_1 + total_monthly_shop_sales_lag_2 + total_monthly_shop_sales_lag_3 + \
        total_monthly_shop_sales_lag_6 + total_monthly_shop_sales_lag_12 + \
        total_monthly_category_sales_lag_1 + \
        total_monthly_shop_category_sales_lag_1 + \
        total_monthly_shop_supercategory_sales_lag_1 + \
        total_monthly_shop_subcategory_sales_lag_1 + \
        total_monthly_city_sales_lag_1 + \
        total_monthly_city_item_sales_lag_1 + \
        total_monthly_supercategory_sales_lag_1 + \
        total_monthly_subcategory_sales_lag_1
    
#date_block_num: 12    =
#shop_id: 2            =
#item_id: 27           =
#item_cnt_month: 0     =
#city_code: 0          =
#item_category_id: 19  =
#type_code: 5          =
#subtype_code: 10      =
```
