```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
train_df = pd.read_csv("sales_train.csv.gz")
test_df  = pd.read_csv("test.csv.gz")
```

```python
monthly = train_df.groupby(["item_id","shop_id","date_block_num"]).item_cnt_day.sum()
monthly = monthly.reset_index()
monthly.columns = ["item_id","shop_id","date_block_num","target"]
```

```python
# January, 2013 is date_block_num==0
from calendar import monthrange

end_date = list()
for block in range(34):
    year  = 2013 + block/12
    month = 1 + block%12
    day   = monthrange(year,month)[1]
    end_date.append( pd.Timestamp("/".join([str(year),str(month),str(day)])) )

blocks = pd.Series(end_date, index=list(range(34)))
 
monthly["end_date"] = monthly["date_block_num"].map(blocks)

monthly.head()
```

```python
series = list()
for name,group in monthly.groupby(["shop_id","item_id"]):
    if len(group) > 10 and group["target"].sum()>100:
        series.append(group)
```

```python
len(series)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

ts = pd.Series(series[2].target.values,index=series[2]["end_date"])
decomposition = seasonal_decompose(ts,freq=12)
```

```python
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ts, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
#plt.subplot(413)
#plt.plot(seasonal,label='Seasonality')
#plt.legend(loc='best')
#plt.subplot(414)
#plt.plot(residual, label='Residuals')
#plt.legend(loc='best')
#plt.tight_layout()
```

```python
per_shop_sales = monthly.groupby(["shop_id","end_date"]).target.sum().reset_index("shop_id")
shop_59 = per_shop_sales[per_shop_sales.shop_id==59]
decomposition = seasonal_decompose(shop_59.target,freq=12)
shop_59
```

```python
trend    = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(shop_59.target, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
from statsmodels.tsa.stattools import acf, pacf

residual.dropna(inplace=True)

lag_acf  = acf(shop_59.target, nlags=20)
lag_pacf = pacf(shop_59.target, nlags=20, method='ols')

plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(shop_59.target, order=(6, 1, 4))  
results_ARIMA = model.fit(disp=-1)
print(results_ARIMA.summary())

predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA = pd.Series(shop_59.target.ix[0], index=shop_59.target.index)
predictions_ARIMA = predictions_ARIMA.add(predictions_ARIMA_diff_cumsum,fill_value=0)
plt.plot(shop_59.target,color='red')
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-shop_59.target)**2)/len(shop_59)))
#plt.plot(shop_59.target)
#plt.plot(results_ARIMA.fittedvalues, color='red')
#plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts_log_diff)**2))
```
